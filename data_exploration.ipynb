## Identify the % of pedestrians involved in major accidents:

```
# Stats vars: 
T = len(  ksi_only[ ksi_only["INJURY"].isin( ["Major",  "Fatal" ] ) ].index )
T_ped = len( 
        ksi_only[ 
                 ksi_only["INVTYPE"].isin( ['Pedestrian'] )  &  ksi_only["INJURY"].isin( ["Major", "Fatal"] ) 
        ].index 
    ) 

### print vars:
print( "Percent pedestrians in major accidents: ", T_ped / T * 100)

```

## Identify the % of accidents happening for pedestrians at intersections

```
#Stats: 
T = len( 
          ksi_only[ 
                   (ksi_only["INVTYPE"].isin( ['Pedestrian'] )) # &  (ksi_only["LOCCOORD"] == 'Mid-Block')
          ].index
) 

T_ped_int = len( 
          ksi_only[ 
                   (ksi_only["INVTYPE"].isin( ['Pedestrian'] ))  &  (ksi_only["LOCCOORD"] == 'Intersection')
          ].index
) 

T_ped_mb = len( 
          ksi_only[ 
                   (ksi_only["INVTYPE"].isin( ['Pedestrian'] ))  &  (ksi_only["LOCCOORD"] == 'Mid-Block')
          ].index
) 
#Print vars 
print( "Total Killed or Seriously Injured Pedestrian Accidents: ", T )
#print( "Num mid block: ", len( ksi_only[ ksi_only["LOCCOORD"] == 'Mid-Block' ].index ) ) 
print("Percent intersection: ", T_ped_int/T * 100  )
print("Percent midblock: ", T_ped_mb/T * 100  )
```
