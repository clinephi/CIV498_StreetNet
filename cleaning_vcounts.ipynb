Our goal is to aggregate these counts by day and sum the daily traffic at each count. Then preferably take only the most recent count for each location. 

```
#1.0 create a day column 
df['date'] = df['time_start'].astype(str).str[0:10]

#2.0 sum all through movements, grouped by mode 
#Through movements 
df["cars_t"] = df["sb_cars_t"] + df["nb_cars_t"] + df["eb_cars_t"] + df["wb_cars_t"]
df["truck_t"] = df["sb_truck_t"] + df["nb_truck_t"] + df["eb_truck_t"] + df["wb_truck_t"] 
df["bus_t"] = df["sb_bus_t"] + df["nb_bus_t"] + df["eb_bus_t"] + df["wb_bus_t"]

#Left turns 
df["cars_l"] = df["sb_cars_l"] + df["nb_cars_l"] + df["eb_cars_l"] + df["wb_cars_l"]
df["truck_l"] = df["sb_truck_l"] + df["nb_truck_l"] + df["eb_truck_l"] + df["wb_truck_l"]
df["bus_l"] = df["sb_bus_l"] + df["nb_bus_l"] + df["eb_bus_l"] + df["wb_bus_l"]

#Right turns
df["cars_r"] = df["sb_cars_r"] + df["nb_cars_r"] + df["eb_cars_r"] + df["wb_cars_r"]
df["truck_r"] = df["sb_truck_r"] + df["nb_truck_r"] + df["eb_truck_r"] + df["wb_truck_r"]
df["bus_r"] = df["sb_bus_r"] + df["nb_bus_r"] + df["eb_bus_r"] + df["wb_bus_r"]

#Peds
df["sum_peds"] = df["nx_peds"] + df["sx_peds"] + df["ex_peds"] + df["wx_peds"]
```

```
#3.0 copy for paranoia's sake
df2 = df.copy( deep=True ) 
```

```
counts_grouped = df2.groupby( ["lat", "lng", "date"])["cars_t", "truck_t", "bus_t", "cars_l", "truck_l", "bus_l", "cars_r", "truck_r", "bus_r", "sum_peds"].apply( lambda x: x.astype( float ).sum() )
```

```
counts_grouped.to_csv( "/content/drive/MyDrive/4.0_Fourth_Year/CIV498/VisionZero/Google Colab/Data_Cleaning/counts_grouped.csv" )
```
